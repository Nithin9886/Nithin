```python
# Standardization (Z-score normalization)
# Formula : Scaled_value = value - min/std

scaled_data = df.copy()

# Volume
sns.distplot(df['Volume'])
```

```python
# Applying standardization formula
for col in ['Volume']:
    scaled_data[col] = (scaled_data[col] - scaled_data[col].mean())/scaled_data[col].std()
sns.distplot(scaled_data['Volume'])
```
